### Meng-import setup environment pandas dan matplotlib

```python
import pandas as pd
from matplotlib import pyplot as plt
```

### Membaca dataset yang telah di download dengan format csv

```python
df = pd.read_csv("Laptop-users.csv")
```

### Menampilkan isi yang ada di dalam dataset

```python
df
```

### Menampilkan type dari atribut yang ada di dalam dataset

```python
df.dtypes
```

### Menampilkan jumlah isi dari atribut income

```python
df['Income'].value_counts()
```

### Membuat tabel untuk atribut income dan membuat atribut jumlah untuk atribut income

```python
labels = df['Income'].value_counts().rename_axis('Income').reset_index(name='Jumlah')
labels
```

### Membuat grafik pie untuk atribut Income

```python
plt.figure(figsize=(8,8), dpi=300)
plt.pie(df['Income'].value_counts(), labels=labels['Income'], autopct='%.1f%%', startangle=90)
plt.title('Income')
plt.xlabel('Jumlah income')
plt.ylabel('Persentase Income')
plt.show()
```

### Menampilkan 5 data paling atas di dalam dataset

```python
df.head()
```

### Membuat Grafik Plot dan Garis untuk atribut Income dan Usia dan hanya menampilkan 5 data

```python
plt.ylabel('Penghasilan')
plt.xlabel('Usia')
plt.title('Grafik garis penghasilan menurut usia')
plt.plot(df.Age[0:5],df.Income[0:5])
plt.show()
```

### Membuat grafik batang untuk membandingkan atribut income menurut usia

```python
x = df['Age']
y = df['Income']

plt.bar(x,y,align = "edge", width = 0.6)
plt.title('perbandingan jumlah income menurut usia')
plt.xlabel('Usia')
plt.ylabel('Jumlah Income')
plt.show()
```
